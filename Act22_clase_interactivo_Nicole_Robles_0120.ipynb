```python
# Definición de la clase Árbitro
class Arbitro:
    def __init__(self, nombre, categoria, experiencia):
        self.nombre = nombre
        self.categoria = categoria
        self.experiencia = experiencia

    def pitar_falta(self):
        print(f"\n[{self.nombre}]: ¡PIIIIII! Se detiene el juego por una infracción.")

    def sacar_tarjeta(self, color, jugador):
        print(f"\n[{self.nombre}]: Muestra tarjeta {color} al jugador {jugador}.")

    def revisar_VAR(self, jugada):
        print(f"\n[{self.nombre}]: Se detiene el juego. Revisando '{jugada}' en el VAR...")
        print(f"[{self.nombre}]: Decisión tomada tras revisar la pantalla.")

# --- INICIO DEL PROGRAMA INTERACTIVO ---

print("=== SISTEMA DE GESTIÓN DE ÁRBITROS ===")
# Captura de datos del usuario
nom = input("Ingrese el nombre del árbitro: ")
cat = input("Ingrese la categoría (ej. Internacional, Local): ")
exp = input("Ingrese los años de experiencia: ")

# Creación del objeto con los datos ingresados
arbitro_usuario = Arbitro(nom, cat, exp)

print(f"\n✅ Árbitro {arbitro_usuario.nombre} registrado con éxito.")

# Menú de acciones
while True:
    print("\n--- MENÚ DE ACCIONES ---")
    print("1. Pitar falta")
    print("2. Sacar tarjeta")
    print("3. Revisar VAR")
    print("4. Salir")

    opcion = input("Seleccione una opción (1-4): ")

    if opcion == "1":
        arbitro_usuario.pitar_falta()
    elif opcion == "2":
        color = input("¿Color de la tarjeta? (Amarilla/Roja): ")
        jugador = input("¿A qué jugador/número?: ")
        arbitro_usuario.sacar_tarjeta(color, jugador)
    elif opcion == "3":
        jugada = input("¿Qué jugada se revisa?: ")
        arbitro_usuario.revisar_VAR(jugada)
    elif opcion == "4":
        print("\nSaliendo del sistema...")
        break
    else:
        print("\nOpción no válida, intenta de nuevo.")

print("\nPrograma finalizado")
print("Creador de software: Nicole Robles NL = 46")
```
